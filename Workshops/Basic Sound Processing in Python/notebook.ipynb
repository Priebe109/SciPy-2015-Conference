```python
import thinkdsp
import thinkplot
import numpy
%matplotlib inline
```

```python
# create a cosine signal (tune c).
cos_sig = thinkdsp.CosSignal(freq=440)
```

```python
# plot the signal.
cos_sig.plot()
thinkplot.config(xlabel='time (s)', legend=False)
```

```python
# make_wave samples the signal at equally-spaced time steps.
wave = cos_sig.make_wave(duration=0.5, framerate=11025)
```

```python
# make audio creates a widget that plays the wave.
wave.apodize()
wave.make_audio()
```

```python
# make_spectrum returns a spectrum object.
spectrum = wave.make_spectrum()
```

```python
spectrum.plot()
thinkplot.config(xlabel='frequency (Hz)', legend=False)
```

```python
# a sawtooth has a more complex harmonic structure.
saw_sig = thinkdsp.SawtoothSignal(freq=440)
saw_sig.plot()
thinkplot.config(xlabel='time (s)', legend=False)
```

```python
saw_wave = saw_sig.make_wave(duration=0.5)
saw_wave.make_audio()
```

```python
# plot the spectrum.
saw_wave.make_spectrum().plot()
thinkplot.config(xlabel='frequency (Hz)', legend=False)
```
